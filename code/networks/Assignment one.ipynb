## Question 1

```python
import networkx as nx
```

```python
def plot_degrees(G, ax, log = False):
        
    degrees = sorted([G.degree(n) for n in G.nodes()], reverse=True)
        
    x = range(1, len(degrees) + 1)
    
    ax.scatter(
        x = np.log(x) if log else x, 
        y = np.log(degrees) if log else degrees, 
        s=5)
        
    return ax
```

```python
n = 1000
m = 50

G = nx.random_graphs.barabasi_albert_graph(n, m, seed=seed)
H = nx.random_graphs.erdos_renyi_graph(n, m/n, seed=seed)

fig, ax = plt.subplots()
ax.set_title("Comparing AB and ER graphs")
ax.set_xlabel("Sorted nodes")
ax.set_ylabel("Degree")

plot_degrees(G, ax, log=True)
plot_degrees(H, ax, log=True)

ax.legend(["Barabàsi - Albert", "Erdos - Rényi"])

fig.show()
```
