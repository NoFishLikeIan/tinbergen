### Core - Periphery model - Brakman et al. 2009



### Model
#### Firm's optimization problem

Let 
\begin{align}
    X := N^{\sigma} \cdot \left[ \frac{1}{N} \cdot \sum^N_{i=1} x_i^{\frac{\epsilon+1}{\epsilon}} \right]^\frac{\epsilon}{\epsilon+1}
\end{align}

then the firm problem is 

\begin{align}
    max_{x_i} \left\{ X \right\} s.t. \sum^N_{i=1} x_i \cdot p_{x_i} = X \cdot P_{X}
\end{align}

yields simple CES function solution

\begin{align} 
    x_i = \left[ \frac{p_{x_i}}{P_X} \right]^{-\epsilon} \cdot X \cdot N^{(\epsilon-1)\cdot(\sigma-1) -1}
\end{align}

where $P_X$ is aggregate as

\begin{align} 
    P_X = N^{1-\sigma}\cdot\left[\frac{1}{N} \sum^N_{i=1} p_{x_i}^{1-\epsilon} \right]^{\frac{1}{1-\epsilon}}
\end{align}


#### Consumption problem

We are going to simplify the problem by making it a two step decision:

- Step 1: $Y$ $\xrightarrow{}$ ($F$ood, $M$anufacturing)
- Step 2: $M \xrightarrow{} c_i$

Based on food $F$ and manufacturing $M$, with Cobb-Douglas $\delta$,

\begin{align}
    F = (1-\delta)\cdot Y, \ M = \delta \frac{Y}{I}
\end{align}

one can once again aggregate to find demand of a given variety,

\begin{align}
    c_i = \frac{p_i^{1 / (\rho - 1)}}{A} \cdot \delta Y, \ A =  \sum p_i^{\rho / (\rho - 1)}
\end{align}

#### Aggregate demand

This implies that, for example, the demand for a variaty in Region 1, with a given price ration

\begin{align}
    \hat{p}_{i, 1} = \frac{p_{i, 1}^{-\epsilon}}{\sum p_{j, 1}^{1-\epsilon}}
\end{align}

\begin{align}
        x_{i, 1} &= \phi \cdot (1-\gamma) \cdot L \cdot \delta w_{F, 1} \cdot \hat{p}_{i, 1} + \lambda \cdot \gamma \cdot L \cdot \delta w_{M, 1} \cdot \hat{p}_{i, 1}\\
    &= \hat{p}_{i, 1} \delta \cdot (\phi \cdot (1-\gamma) \cdot L \cdot w_{F, 1} + \lambda \cdot \gamma \cdot L \cdot w_{M, 1}) \\
    &= \hat{p}_{i, 1} \cdot \delta Y_1 
\end{align}

#### Price setting

With transport cost $T$ and wage $w_{k}$, the profits of a firm in region $k$ producing good $i$ are

\begin{align}
    \pi_{i \vert k} = (p_{i, k} - \beta w_{k}) \cdot x_{i, k} + (p_{i, -k} - T \cdot \beta w_{k}) \cdot x_{i, -k}) - \alpha w_{k}
\end{align}

given the monopolistic competition the prices are

\begin{align}
    p_{i, k} = \beta \cdot w_k \cdot \frac{\epsilon}{1-\epsilon} = \beta \cdot w_k \cdot \rho \implies p_{i, -k} = T\cdot p_{i, k} 
\end{align}

#### Solution

Using the above prices and the zero profit condition

\begin{align}
     \pi_{i, k} &= \left(\frac{1}{\rho} - 1 \right) \cdot x_i \cdot \beta \cdot w_k - \alpha w_k = 0 \\
     \implies x_i &= \frac{\alpha}{\beta} \cdot (\epsilon - 1)
\end{align}

In the job market the labour demand is then $N \cdot \alpha \epsilon = \gamma L$

### Equation system

\begin{align}
    Y_1 &= \phi \cdot (1-\gamma)L + \lambda \cdot \gamma L \cdot w \\
    Y_2 &= (1-\phi) \cdot (1-\gamma)L + (1-\lambda) \cdot \gamma L \cdot w
\end{align}
