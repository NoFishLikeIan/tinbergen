```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
lambdas = np.linspace(0, 2*np.pi, 200)
```

```python
phi_2 = -1/4
thetas = np.linspace(0, np.pi)
phi_ones = [np.cos(theta) for theta in thetas]
```

```python
def make_density(phi_1):
    def density(lmbd):
        den = 1 + np.square(phi_2) + np.square(phi_1) - 2 * \
            phi_1*(1-phi_2)*np.cos(lmbd) - 2*phi_2*np.cos(2*lmbd)

        return 1 / (2*np.pi*den)

    return density
```

```python
plt.figure(figsize=(12, 8))

for n, phi_1 in enumerate(phi_ones):
    print(f'{n}/{len(phi_ones)}', end='\r')
    density = make_density(phi_1)
    dens = [density(l) for l in lambdas]
    plt.xlabel('Lambda')
    plt.ylabel('Density')
    plt.plot(lambdas, dens)

plt.show()
```

```python
def density_ar_4(lmb, phi_4):
    den = np.square(1 - phi_4*np.cos(4*lmb)) + phi_4 * np.square(np.sin(4*lmb))
    
    return 1 / (den* 2*np.pi)
```

```python
dens_sim = [density_ar_4(l, 0.99) for l in lambdas]
#dens_sim_2 = [density_ar_4(l, .02) for l in lambdas]


plt.figure(figsize=(12, 8))

plt.plot(lambdas, dens_sim)
#plt.plot(lambdas, dens_sim_2)

v = 0
for i in range(4):
    plt.axvline(v, c='red', alpha=0.5)
    v += np.pi/2

plt.xlabel('Lambda')
plt.ylabel('Density')
plt.show()
```
